```python
import random
import matplotlib.pyplot as plt

def simular_embarque():
    asientos = [False] * 100 
    asiento_elegido = random.randint(0, 99)
    asientos[asiento_elegido] = True


    for pasajero in range(1, 99):  
        if not asientos[pasajero]:  
            asientos[pasajero] = True
        else:  
            asientos_libres = [i for i, libre in enumerate(asientos) if not libre]
            asiento_elegido = random.choice(asientos_libres)
            asientos[asiento_elegido] = True
    return asientos[99]  

num_simulaciones = 100000
exitos = sum(simular_embarque() for _ in range(num_simulaciones))
probabilidad = exitos / num_simulaciones

plt.hist([simular_embarque() for _ in range(1000)], bins=[0, 1/2, 1], edgecolor='black')
plt.xticks([0.25, 0.75], ['Asiento 1', 'Asiento 100'])
plt.ylabel('Frecuencia')
plt.title('Asiento final del pasajero 100 (1000 simulaciones)')
plt.show()
```
